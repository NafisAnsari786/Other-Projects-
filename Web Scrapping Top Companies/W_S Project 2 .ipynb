#  Web Scrapping Top 100 Indian Companies 

```python
from bs4 import BeautifulSoup
import requests as r
```

```python
url='https://www.capitalmarket.com/Compendium/Ranking-Top-500-companies'

page=r.get(url)
page
```

```python
soup=BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
table = soup.find_all('table')[1:]
print(table)
```

```python
companies1=soup.find_all('th')[:2]
companies1
```

```python
comp_titles=[title.text.strip() for title in companies1]
comp_titles
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(columns=comp_titles)
df
```

```python
col_data=soup.find_all('tr')
```

```python
for row in col_data[3:53]:
    row_data=row.find_all('td')
    # print(row_data)
    each_row_data=[data.text.strip() for data in row_data]

    length=len(df)
    df.loc[length]=each_row_data
```

```python
df
```

```python
for row in col_data[55:104]:
    row_data=row.find_all('td')
    # print(row_data)
    each_row_data=[data.text.strip() for data in row_data]

    length=len(df)
    df.loc[length]=each_row_data
```
